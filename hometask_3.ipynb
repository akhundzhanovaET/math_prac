Plot the populations for the years in the period. Add legend and axis labels to the plot.

Find which year each species had the largest population. Mark these years in the plot of populations.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('http://scipy-lectures.org/_downloads/populations.txt')
years = data[:,0]
hares = data[:, 1]
lynxes = data[:, 2]

plt.plot(years, hares, label = 'Hares', c = 'blue')
plt.plot(years, lynxes, label = 'Lynxes', c = 'orange')

plt.axvline(years[np.argmax(hares)], c = 'blue')
plt.axvline(years[np.argmax(lynxes)], c = 'orange')

plt.xticks([years[np.argmax(hares)],years[np.argmax(lynxes)]],rotation = 90)
plt.xlabel('YEARS')
plt.ylabel('POPULATION')
plt.legend()
plt.show()
```

Find which species (hares or lynxes) has the largest population for each year. Print the result as [H, H, L, H, ...]. Plot a pie chart showing ratio of "H" and "L" values obtained.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('http://scipy-lectures.org/_downloads/populations.txt')
years = data[:,0]
hares = data[:,1]
lynxes = data[:,2]

plt.pie([sum(lynxes >= hares),sum(hares > lynxes)] , labels = ['Lynxes','Hares'])
plt.show()
a = list()
for i in range(len(years)):
    if lynxes[i] >= hares[i]:
        a.append('L')
    else:
        a.append('H')
print(a)
  
```

Find which years any of the populations is above 50000. Show time periods where populations are above 50000 in the plot of populations.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('http://scipy-lectures.org/_downloads/populations.txt')
years = data[:,0]
hares = data[:,1]
lynxes = data[:,2]
carrots = data[:,3]
plt.plot(years, hares, label = 'Hares')
plt.plot(years, lynxes, label = 'Lynxes')
plt.axhline(y=50000, color='r', linestyle="dashed")
a = np.maximum(hares,lynxes)
plt.xticks(years[a > 50000],rotation = 90)
plt.xlabel('YEARS')
plt.ylabel('POPULATION')
plt.legend()
plt.show()
res = {'hares': data[hares.argsort()[:2], 0],
        'lynxes': data[lynxes.argsort()[:2], 0],
       'carrot' : data[carrots.argsort()[:2], 0]}
print(res)
```

Show population of hares vs carrots and hares vs lynxes.

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('http://scipy-lectures.org/_downloads/populations.txt')
years = data[:,0]
hares = data[:,1]
lynxes = data[:,2]
carrots = data[:,3]

fig,ax = plt.subplots(1,2)
summa = hares+carrots
ax[0].bar(years,hares/summa)
ax[0].bar(years,carrots/summa,bottom = hares/summa)
ax[0].set_title('Hares(b) vs Carrots(or)')
summa = hares+lynxes
ax[1].bar(years,hares/summa)
ax[1].bar(years,lynxes/summa,bottom = hares/summa)
ax[1].set_title('Hares(b) vs Lynxes(or)')

plt.show()
```

Find a number of principal components for 99%, 95%, 90%, and 85% of explained variance.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
digits_new = fetch_openml('mnist_784')
pca = PCA()

X_transformed = pca.fit_transform(digits_new.data.astype(float)) 
var = pca.explained_variance_ratio_.cumsum().round(2) * 100
plt.plot(var)
plt.ylabel('total varience')
plt.xlabel('number of components')
plt.show()
```
