Plot the populations for the years in the period. Add legend and axis labels to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('http://scipy-lectures.org/_downloads/populations.txt')
years = data[:,0]
hares = data[:,1]
lynxes = data[:,2]
carrots = data[:,3]

plt.plot(years, hares, label = 'Hares')
plt.plot(years, lynxes, label = 'Lynxes')
plt.plot(years, carrots, label = 'Carrots')
plt.xticks(np.arange(min(years), max(years) + 1, 5.0))
plt.xlabel('YEARS')
plt.ylabel('POPULATION')
plt.legend()
plt.show()

```

Assuming total population of hares and lynxes is 100%, plot stacked bar graph showing ratio of each specis for the years in the period.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('http://scipy-lectures.org/_downloads/populations.txt')
years = data[:,0]
total = data[:,1] + data[:,2]
hares_ratio = data[:,1]/total
lynxes_ratio = data[:,2]/total

plt.bar(years,hares_ratio)
plt.bar(years,lynxes_ratio,bottom = hares_ratio)
plt.xlabel('YEARS')
plt.ylabel('RATIO')
plt.show()

```

Find the mean and std of the populations of each species. Plot the historgam of population for each species and show mean values with vertical line. Arrange 3 subplots in a row.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('http://scipy-lectures.org/_downloads/populations.txt')
fig, ax = plt.subplots(1,3) #     ,     

ax[0].hist(data[:, 1], bins = 15)
ax[0].axvline(data[:,1].mean(), c = "r")

ax[1].hist(data[:, 2], bins = 15)
ax[1].axvline(data[:,2].mean(), c = "r")

ax[2].hist(data[:, 3], bins = 15)
ax[2].axvline(data[:,3].mean(), c = "r")

plt.show()
```

Find which year each species had the largest population. Mark these years in the plot of populations.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('http://scipy-lectures.org/_downloads/populations.txt')
years = data[:,0]
hares = data[:, 1]
lynxes = data[:, 2]

plt.plot(years, hares, label = 'Hares')
plt.plot(years, lynxes, label = 'Lynxes')

plt.axvline(years[np.argmax(hares)], c = 'r')
plt.axvline(years[np.argmax(lynxes)], c = 'r')

plt.xticks([years[np.argmax(hares)],years[np.argmax(lynxes)]],rotation = 90)
plt.xlabel('YEARS')
plt.ylabel('POPULATION')
plt.legend()
plt.show()
```

Find which species (hares or lynxes) has the largest population for each year. Print the result as [H, H, L, H, ...]. Plot a pie chart showing ratio of "H" and "L" values obtained.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('http://scipy-lectures.org/_downloads/populations.txt')
years = data[:,0]
hares = data[:,1]
lynxes = data[:,2]

plt.pie([sum(lynxes >= hares),sum(hares > lynxes)] , labels = ['Lynxes','Hares'])
plt.show()
a = list()
for i in range(len(years)):
    if lynxes[i] >= hares[i]:
        a.append('L')
    else:
        a.append('H')
print(a)
  
```

Find which years any of the populations is above 50000. Show time periods where populations are above 50000 in the plot of populations.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('http://scipy-lectures.org/_downloads/populations.txt')
years = data[:,0]
hares = data[:,1]
lynxes = data[:,2]

plt.plot(years, hares, label = 'Hares')
plt.plot(years, lynxes, label = 'Lynxes')

a = np.maximum(hares,lynxes)
plt.xticks(years[a > 50000],rotation = 90)
plt.xlabel('YEARS')
plt.ylabel('POPULATION')
plt.legend()
plt.show()
```

Find the top 2 years for each species when they had the lowest populations.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('http://scipy-lectures.org/_downloads/populations.txt')
years = data[:,0]
hares = data[:,1]
lynxes = data[:,2]

haresYear1 = np.argmin(hares)
haresYear2 = np.argmin(np.insert(np.delete(hares,haresYear1), haresYear1,10 ** 10))
lynxesYear1 = np.argmin(lynxes)
lynxesYear2 = np.argmin(np.insert(np.delete(lynxes,lynxesYear1), lynxesYear1,10 ** 10))

plt.plot(years, hares, label = 'Hares')
plt.plot(years, lynxes, label = 'Lynxes')
plt.xticks([years[haresYear1],years[haresYear2],years[lynxesYear1],years[lynxesYear2]],rotation = 90)
plt.legend()
plt.show()

```
