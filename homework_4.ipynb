```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url1 = ('https://raw.githubusercontent.com/vaibhavwalvekar/NYC-Flights-2013-Dataset-Analysis/master/flights.csv')
flight = pd.read_csv(url1)

url2 = ('https://raw.githubusercontent.com/vaibhavwalvekar/NYC-Flights-2013-Dataset-Analysis/master/weather.csv')
weather =  pd.read_csv(url2)
```

Сколько рейсов приходится на каждый аэропорт?

```python
flight_sum=flight.origin.value_counts().reset_index(name='NumberOfFlights')
flight_sum
```

```python
plt.pie(flight_sum.NumberOfFlights,labels=['EWR','JFK','LGA'],autopct='%1.1f%%')
plt.show()
```

Во все аэропорты прибывает примерно одинаковое число рейсов. Значит их можно считать достаточно равнозначными.

Посмотрим на соотношение задержек в каждом аэропорту

```python
origin_delay=pd.DataFrame(flight[flight.arr_delay>0].groupby(['origin'])['arr_delay'].sum().reset_index(name='TotalDelay'))
origin_delay
```

```python
plt.pie(origin_delay.TotalDelay, labels=['EWR','JFK','LGA'], autopct='%1.1f%%')
plt.show()
```

Как видим, наибольшее количество задержки приходится на EWR. Это нужно учитывать при оптимизации и организации новых/других самолетов и пассажиров.

Посмотрим, влияет ли и как длительность полета на задержки в каждый из трех аэропортов.

```python
flight['flight_delay']=flight['arr_delay']-flight['dep_delay']

lengthEWR_1=flight.loc[(flight['origin'] == 'EWR') & (flight['air_time']<500)].groupby(['air_time'])['flight_delay'].mean()
lengthEWR_2=flight.loc[(flight['origin'] == 'EWR') & (flight['air_time']>500)].groupby(['air_time'])['flight_delay'].mean()

lengthLGA_1=flight.loc[(flight['origin'] == 'LGA') & (flight['air_time']<500)].groupby(['air_time'])['flight_delay'].mean()
lengthLGA_2=flight.loc[(flight['origin'] == 'LGA') & (flight['air_time']>500)].groupby(['air_time'])['flight_delay'].mean()

lengthJFK_1=flight.loc[(flight['origin'] == 'JFK') & (flight['air_time']<500)].groupby(['air_time'])['flight_delay'].mean()
lengthJFK_2=flight.loc[(flight['origin'] == 'JFK') & (flight['air_time']>500)].groupby(['air_time'])['flight_delay'].mean()

ax_lengthEWR_1 = lengthEWR_1.plot(label="задержка")
ax_lengthEWR_1.set_title('Зависимость задержки от длительности полета из аэропорта EWR (менее 500 минут полета)',loc = 'left')
ax_lengthEWR_1.set_xlabel('время в воздухе (в минутах)')
ax_lengthEWR_1.set_ylabel('среднее время задержки')
plt.legend()
plt.show()

ax_lengthEWR_2 = lengthEWR_2.plot(label="задержка")
ax_lengthEWR_2.set_title('Зависимость задержки от длительности полета из аэропорта EWR (более 500 минут полета)',loc = 'left')
ax_lengthEWR_2.set_xlabel('время в воздухе (в минутах)')
ax_lengthEWR_2.set_ylabel('среднее время задержки')
plt.legend()
plt.show()

ax_lengthLGA_1 = lengthLGA_1.plot(label="задержка")
ax_lengthLGA_1.set_title('Зависимость задержки от длительности полета из аэропорта LGA (менее 500 минут полёта)',loc = 'left')
ax_lengthLGA_1.set_xlabel('время в воздухе (в минутах)')
ax_lengthLGA_1.set_ylabel('среднее время задержки')
plt.legend()  
plt.show()

ax_lengthLGA_2 = lengthLGA_2.plot(label="задержка")
ax_lengthLGA_2.set_title('Зависимость задержки от длительности полета из аэропорта LGA (более 500 минут полёта)',loc = 'left')
ax_lengthLGA_2.set_xlabel('время в воздухе (в минутах)')
ax_lengthLGA_2.set_ylabel('среднее время задержки')
plt.legend()  
plt.show()

ax_lengthJFK_1 = lengthJFK_1.plot(label="задержка")
ax_lengthJFK_1.set_title('Зависимость задержки от длительности полета из аэропорта JFK (менее 500 минут полета)',loc = 'left')
ax_lengthJFK_1.set_xlabel('время в воздухе (в минутах)')
ax_lengthJFK_1.set_ylabel('среднее время задержки')
plt.legend()
plt.show()

ax_lengthJFK_2 = lengthJFK_2.plot(label="задержка")
ax_lengthJFK_2.set_title('Зависимость задержки от длительности полета из аэропорта JFK (более 500 минут полета)',loc = 'left')
ax_lengthJFK_2.set_xlabel('время в воздухе (в минутах)')
ax_lengthJFK_2.set_ylabel('среднее время задержки')
plt.legend()
plt.show()
```

В аэропорт LGA вообще не прибывали самолеты с длительностью полета более 500 минут. Начиная с примерно 400 минут длительности полета увеличивается задержка. Это нужно учитывать при оптимизации и организации новых/других самолетов и пассажиров. При этом в JFK самолеты с большой длительностью полета имеют меньшую задержку в сравнении с другими аэропортами. Как вариант тогда можно ставить прибытие самолетов с большой длительностью полета в JFK

Посмотрим, как скорость ветра влияет на задержки

```python
flights=flight.merge(weather,on=['year','month', 'day', 'hour', 'origin'], how='inner')
wind=pd.DataFrame(flights[flights.dep_delay>0].groupby(['wind_speed'])['dep_delay'].mean().reset_index(name='mean_delay'))
wind=wind.sort_values("wind_speed")
wind=wind.drop(labels=[wind.index[-1]])
```

```python
plt.figure(figsize=(10,5))
plt.xticks(wind.wind_speed)
plt.xticks(rotation=90)
plt.plot(wind.wind_speed, wind.mean_delay)
plt.grid()
plt.xlabel('скорость ветра')
plt.ylabel('среднее время задержки')
plt.show()
```

Скорость ветра и среднее время задержки прямо пропорциональны (сильнее ветер - больше время задержки(при полете против ветра))

Посмотрим, как порывы ветра влияют на задержки

```python
gust=pd.DataFrame(flights[flights.arr_delay>0].groupby(['wind_gust'])['arr_delay'].mean().reset_index(name='mean_delay'))
gust=gust.drop(labels=[gust.index[-1]])
```

```python
plt.figure(figsize=(20,5))
plt.scatter(gust.wind_gust,gust.mean_delay)
plt.xticks(rotation=90)
plt.xticks(gust.wind_gust)
plt.ylabel('среднее время задержки')
plt.xlabel('порывы ветра')
plt.grid()
plt.show()
```

Задержки значительно больше при порывах ветра больше 30

Какие еще природные факторы влияют на задержку?

```python
temperature=flights.groupby(['temp'])['dep_delay'].mean().to_frame().reset_index()
ax_temperature=temperature.plot.scatter('temp', 'dep_delay')
ax_temperature.set_xlabel('температура')
ax_temperature.set_ylabel('среднее время задержки (в минутах)')
plt.show()

dewp=flights.groupby(['dewp'])['dep_delay'].mean().to_frame().reset_index()
ax_dewp=dewp.plot.scatter('dewp', 'dep_delay')
ax_dewp.set_xlabel('роса')
ax_dewp.set_ylabel('среднее время задержки (в минутах)')
plt.show()

humidity=flights.groupby(['humid'])['dep_delay'].mean().to_frame().reset_index()
ax_humidity=humidity.plot.scatter('humid', 'dep_delay')
ax_humidity.set_xlabel('влажность')
ax_humidity.set_ylabel('среднее время задержки (в минутах)')
plt.show()

pressure=flights.groupby(['pressure'])['dep_delay'].mean().to_frame().reset_index()
ax_pressure=pressure.plot.scatter('pressure', 'dep_delay')
ax_pressure.set_xlabel('давление')
ax_pressure.set_ylabel('среднее время задержки (в минутах)')
plt.show()
```

Среднее время задержки увеличивается из-за низкого давления, слишком большого или наоборот слишком низкого значения росы и высокой температуры. Влажность несильно влияет на задержки.

Как время суток влияет на задержки?

```python
hours=pd.DataFrame(flights[flights.arr_delay>0].groupby(['hour'])['arr_delay'].mean().reset_index(name='mean_delay'))
```

```python
plt.figure(figsize=(10,5))
plt.plot(hours.hour,hours.mean_delay)
plt.xticks(hours.hour)
plt.ylabel('среднее время задержки')
plt.xlabel('время суток')
plt.grid()
plt.show()
```

Самое большое среднее время задержек между 23:00 и 2:00

В какие месяцы наибольшие задержки?

```python
month = flights[flights.arr_delay>0].groupby(['month'])['dep_delay'].mean()
ax_month = month.plot.pie(label='номер месяца',autopct='%1.1f%%')
plt.show()
```

Наибольшие задержки в теплые месяцы и в декабрь.

В какие месяцы наибольшая загруженность?

```python
ewr=flight.loc[flight['origin'] == 'EWR'].groupby(['month'])['Unnamed: 0'].sum()
jfk=flight.loc[flight['origin'] == 'JFK'].groupby(['month'])['Unnamed: 0'].sum()
lga=flight.loc[flight['origin'] == 'LGA'].groupby(['month'])['Unnamed: 0'].sum()
ewr.plot.pie(label='EWR',autopct='%1.1f%%')
plt.show()
jfk.plot.pie(label='JFK',autopct='%1.1f%%')
plt.show()
lga.plot.pie(label='LGA',autopct='%1.1f%%')
plt.show()
```

Аэропорты гораздо загруженнее в теплое время года (многие в это время летят в отпуск). Учитывая предыдущий график, что в теплое время наибольшие задержки, следует, например, увеличить количество персонала. 

Какие перевозчики чаще всего задерживаются?

```python
carriers=pd.DataFrame(flights[flights.arr_delay>0].groupby(['carrier'])['arr_delay'].mean().reset_index(name='mean_delay'))
plt.figure(figsize=(10,5))
plt.bar(carriers.carrier, carriers.mean_delay)
plt.xticks(carriers.carrier)
plt.xlabel('перевозчик')
plt.ylabel('среднее время задержки')
plt.show()
```
